```python
class TreeNode:

    def __init__(self, value, left_child=None, right_child=None):
        self.value = value              # string, stored value
        self.left_child = left_child    # None or TreeNode instance
        self.right_child = right_child  # None or TreeNode instance

    def get_value(self):
        return self.value

    def add_left_child(self, value):
        pass

    def add_right_child(self, value):
        pass

    def get_left_child(self):
        return self.left_child

    def get_right_child(self):
        return self.right_child

    def has_left_child(self) -> bool:
        if self.left_child != None:
            return True
        else:
            return False

    def has_right_child(self) -> bool:
        if self.right_child != None:
            return True
        else:
            return False

    def is_leaf(self) -> bool:
        if self.has_left_child() or self.has_right_child():
            return False
        else:
            return True

    def __str__(self):
        pieces = []
        self._parenthesize_recur(pieces)
        return ''.join(pieces)
    
    def _parenthesize_recur(self, result):
        if self.is_leaf():
            if self.value == None:
                print(self.value)
            else:
                result.append(str(self.value))
        else:
            #tylko do naszego zadania
            if self.value in ['sin','cos','exp','log']:
                result.append(str(self.value))
                result.append("(")
                if self.has_right_child():
                    self.right_child._parenthesize_recur(result)
                result.append(")")
            else:    
                result.append("(")
                if self.has_left_child():
                    self.left_child._parenthesize_recur(result)
                result.append(str(self.value))
                if self.has_right_child():
                    self.right_child._parenthesize_recur(result)
                result.append(")")
```

```python
funcs = ['sin','cos','exp','log']

def tokenize(raw):
    SYMBOLS = set('+-*^/() ') # we do not allow x as multiplication
    mark = 0
    tokens = []
    n = len(raw)
    for j in range(n):
        if raw[j] in SYMBOLS:
            if mark != j:
                tokens.append(raw[mark:j]) # complete preceding token
            if raw[j] != ' ':
                tokens.append(raw[j]) # include this token
            mark = j+1 # update mark to being at next index
    if mark != n:
        tokens.append(raw[mark:n]) # complete preceding token
    return tokens

def build_expression_tree(tokens):
    S = [] # we use Python list as stack
    for t in tokens:
        if t in '+-*/' or t in funcs: # t is an operator symbol
            S.append(t) # push the operator symbol
        elif t not in '()': # consider t to be a literal
            S.append(TreeNode(t)) # push trivial tree storing value
        elif t == ')': # compose a new tree from three constituent parts
            right = S.pop() # right subtree as per LIFO
            op = S.pop() # operator symbol
            ### what to do with left sub tree for sin cos log exp?
            if op not in funcs:    
                left = S.pop() # left subtree
                S.append(TreeNode(op, left, right)) # repush tree
            else:
                S.append(TreeNode(op, left_child =TreeNode(None) ,right_child=right)) # repush tree
        # we ignore a left parenthesis
    return S.pop()
```

```python
expr = "exp((x+1))"
tokens = tokenize(expr)
print(tokens)
exprtree = build_expression_tree(tokens)
print(exprtree)
print(exprtree.value in ['sin','cos','exp','log'])


#testy dla innych wyrażeń
print(build_expression_tree(tokenize("(x+1)")))
print(build_expression_tree(tokenize("(x*1)")))



```
