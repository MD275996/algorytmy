```python
import time
import matplotlib.pyplot as plt
import numpy
import math
from scipy.optimize import curve_fit
```

```python
def sequence(n):
    return [i for i in range(-math.ceil(n),0)]

def gettime(s):
    t = time.time()
    sorted(s)
    t = time.time() - t
    return t

def nlogn(s):
    tab = []
    for el in s:
        tab.append(el*math.log(el))
    return tab

def func(x,a):
    return a*x*numpy.log(x)

def fun(x,a,b):
    return a*x+b
```

```python
n = numpy.linspace(10000,50000000,10)
times = [gettime(sequence(el)) for el in n]
```

```python
x_rang = numpy.arange(n[0],n[-1],100)
popt, pcov = curve_fit(func,n,times)
popt2, pcov2 = curve_fit(fun,n,times)

plt.plot(n,times, "ro",label = "dane")
plt.plot(x_rang,func(x_rang,*popt),label = "nlogn")
#plt.plot(x_rang,fun(x_rang,*popt2),label = "n")
plt.xlabel("Ilość danych")
plt.ylabel("Czas")
plt.legend(loc='upper left')
plt.title("Analiza złożoności")
plt.show()

```
