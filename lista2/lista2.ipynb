```python
def sequence(n):
    """funkcja która zwraca tablice o rozmiarze n zawierająca elementy z przedziału [-100,100]"""
    return [random.randint(-100,100) for i in range(n)]

def recurence_search(an,max):
    """
    Opis:  
        Funkcja rekurencyjna wyznaczająca największy element sekwencji an, poprzez wycinanie jednego elementu i
        porównywanie go z obecnym max

    Argumenty:
        an : lista
            sekwencja losowych liczb, nieposortowana

        max : int
            obecna największa wartość, przy pierwszym wywołaniu funkcji przypisana jest wartość an[1]

    Zwraca:
        największą liczbę z sekwencji
    
    """  
    if len(an)==0:
        return max
    else:
        a = an.pop()
        if a > max:
            max = a
        return recurence_search(an,max)
```

Łatwo zauważyć że funkcja będzie miała złożoność liniową, gdyż dla każdego elementu tablicy wykonuje stałą liczbę operacji

```python
# an = sequence(10)
an = [100]
print(an)
max = recurence_search(an,an[0])
print(max)
```

<h2>Zadanie trzecie</h2>

```python
def minmax(an,min,max):
    """
    Opis:  
        Funkcja rekurencyjna wyznaczająca największy i najmniejszy element sekwencji an, poprzez wycinanie jednego elementu i
        porównywanie go z obecnym min i max

    Argumenty:
        an : lista
            sekwencja losowych liczb, nieposortowana

        min : int
            obecna najmniejsza wartość, przy pierwszym wywołaniu funkcji przypisana jest wartość an[1]

        max : int
            obecna największa wartość, przy pierwszym wywołaniu funkcji przypisana jest wartość an[1]

    Zwraca:
        parę liczb, gdzie pierwsza jest najmniejszą z sekwencji, druga największą
    
    """  
    if len(an)==0:
        return min,max
    else:
        a = an.pop()
        if a > max:
            max = a
        elif a < min:
            min = a
        return minmax(an,min,max)
```

```python
an = [100]
print(an)
min,max = minmax(an,an[0],an[0])
print(min,max)
```

<h2>Zadanie piąte</h2>

```python
def palindromcheck(str1):
    """
    Opis:  
        Funkcja rekurencyjna sprawdzająca czy dany ciąg znaków jest palindromem

    Argumenty:
        str: string
            ciąg znaków do sprawdzenia

    Zwraca:
        bool True jeśli jest palindromem, False jeśli nie jest
    
    """ 
    if len(str1) <= 1:
        return True
    if str1[0] == str1[-1]:
        return palindromcheck(str1[1:][:-1])
    else:
        return False
```

```python
print(palindromcheck("kajak"))
print(palindromcheck("kamilślimak"))
print(palindromcheck("kot"))
```

```python
def find(path,filename):
    """
    Opis:  
        Funkcja rekurencyjna która odnajduje pliki o nazwie filename i wypisuje ścieżki znalezionych plików na ekran

    Argumenty:
        path : string
            ścieżka od której startujemy poszukiwania

        filename : string
            nazwa plików które mają być wyszukane

    Zwraca:
        -
    
    """ 
    for el in os.listdir(path):
        if isfile(join(path,el)):
            if el[:-4] == filename:
                print(join(path,el))
        else:
            find(join(path,el),filename)


def find2(path,filename):
    """
    Opis:  
        Funkcja rekurencyjna która odnajduje pliki o nazwie filename i zwraca listę ich ścieżek

    Argumenty:
        path : string
            ścieżka od której startujemy poszukiwania

        filename : string
            nazwa plików które mają być wyszukane

    Zwraca:
        listę ścieżek odnalezionych plików
    
    """

    elements_tab=[]
    for i in range(len(path)):
        #print(os.listdir(path[i]))
        for el in os.listdir(path[i]):
            #print(el)
            #print(path[i])
            elements_tab.append(join(path[i],el))
    
    files=[]
    dirs=[]
    for el in elements_tab:
        if isfile(el):
            if os.path.basename(el)[:-4] == filename:
                files.append(join(path[0],el))
        else:
            dirs.append(join(path[0],el))
    if len(dirs) != 0:
        return files + find2(dirs,filename)
    else:
        return files 
```
